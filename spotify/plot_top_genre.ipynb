```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Read the CSV file
df = pd.read_csv('top_counties_with_genres.csv')

# Create the plot
plt.figure(figsize=(15, 8))
sns.barplot(x='GeoName', y='CombinedScore', data=df, palette='viridis')

# Customize the plot
plt.title('Top Counties by Combined Score with Top Genre', fontsize=18)
plt.xlabel('County', fontsize=14)
# plt.ylabel('Combined Score', fontsize=14)
plt.xticks(rotation=90, ha='right')
plt.ylim(0, 15)
plt.yticks([])

# Add genre labels to the bars
for i, row in df.iterrows():
    plt.text(i, row['CombinedScore'], row['TopGenre'], 
             ha='center', va='bottom', rotation=90, fontsize=18)

plt.tight_layout()
# plt.savefig('top_counties_genre_plot.png', dpi=300, bbox_inches='tight')
# plt.close()
plt.show()

print("Bar plot saved as 'top_counties_genre_plot.png'")

# Print summary
print("\nTop Counties with Their Top Genre:")
for _, row in df.iterrows():
    print(f"{row['GeoName']}: {row['TopGenre']} (Score: {row['CombinedScore']:.2f})")
```
