```python
import json
import pandas as pd
import os
import numpy as np
import requests
from dotenv import load_dotenv
import pathlib as path
```

```python
load_dotenv()
api_key = os.getenv("CENSUS_API_KEY")
```

```python
#2010 State Populations

state2010_url = f'https://api.census.gov/data/2010/dec/sf1?get=P001001,NAME&for=state:*&key={api_key}'

state2010_query = requests.get(state2010_url).json()
print(json.dumps(state2010_query, indent=4))
```

```python
state_pop_df = pd.DataFrame(state2010_query)
state_pop_df.head()

state_pop_df.columns = state_pop_df.iloc[0]
state_pop_df = state_pop_df[1:]
state_pop_df.reset_index(drop=True, inplace=True)
state_pop_df.head()

state_pop_df = state_pop_df[["NAME", "P001001", "state"]]
state_pop_df2 = state_pop_df.rename(columns={"NAME": "Name", "P001001": "Total Population", "state": "State Code"})
state_pop_df2.head()
#query_df.describe()
```
