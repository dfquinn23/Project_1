```python
import json
import pandas as pd
import os
import numpy as np
import requests
from dotenv import load_dotenv
import pathlib as path
```

```python
load_dotenv()
api_key = os.getenv("CENSUS_API_KEY")
```

```python
#Total # of housing units in the US in 2010

housing2010_query_url = f'https://api.census.gov/data/2010/dec/sf1?get=NAME,H001001&for=state:*&key={api_key}'
query = requests.get(housing2010_query_url).json()
print(json.dumps(query, indent=4))
```

```python
housing2010_query_df = pd.DataFrame(query)
housing2010_query_df.head()
#query_df.describe()
```

```python
#Total Population 2010

state2010_query_url = f'https://api.census.gov/data/2010/dec/sf1?get=NAME,P001001&for=state:*&key={api_key}'

query = requests.get(state2010_query_url).json()
print(json.dumps(query, indent=4))
```

```python
state2010_query_df = pd.DataFrame(query)
state2010_query_df.head()
#query_df.describe()
```

```python
county2010_query_url = f'https://api.census.gov/data/2020/dec/pl?get=NAME,P1_001N&for=county:*&in=state:*&key={api_key}'

query = requests.get(county2010_query_url).json()
print(json.dumps(query, indent=4))
```

```python
county2010_query_df = pd.DataFrame(query)
county2010_query_df
#query_df.describe()
```
