```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Read the data
data = pd.read_csv('data/top_counties.csv')

# # Rename '2022' column to 'Income Growth'
# data = data.rename(columns={'2022': 'Income Growth'})

plt.figure(figsize=(20, 10))

pos = np.arange(len(data))
width = 0.25

plt.bar(pos, data['Income Growth'], width, alpha=0.8, color='#FFA07A', label='Income Growth')
plt.bar(pos + width, data['CAGR'], width, alpha=0.8, color='#98FB98', label='CAGR')
plt.bar(pos + 2*width, data['Total Growth Rate'], width, alpha=0.8, color='#87CEFA', label='Total Growth Rate')

plt.xlabel('Location', fontsize=18)
plt.ylabel('Rate', fontsize=18)
plt.ylim(0, 50)
plt.title('Growth Rates by Location', fontsize=22)
plt.xticks(pos + width, data['GeoName'], rotation=90, ha='right', fontsize=18)
plt.yticks(fontsize=18)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=18)

# Add value labels on top of each bar
for i in range(len(data)):
    plt.text(pos[i], data['Income Growth'][i], f'{data["Income Growth"][i]:.2f}', 
             ha='center', va='bottom', rotation=90, fontsize=12)
    plt.text(pos[i] + width, data['CAGR'][i], f'{data["CAGR"][i]:.2f}', 
             ha='center', va='bottom', rotation=90, fontsize=14)
    plt.text(pos[i] + 2*width, data['Total Growth Rate'][i], f'{data["Total Growth Rate"][i]:.2f}', 
             ha='center', va='bottom', rotation=90, fontsize=14)

plt.tight_layout()

# plt.savefig("growth_rates_comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```
